## 3. Rearrange and rename the columns for consistency across all files

```python
file2 = file2[["Customer", "ST", "GENDER", "Education", "Customer Lifetime Value", "Income", "Monthly Premium Auto", "Number of Open Complaints", "Policy Type", "Vehicle Class", "Total Claim Amount"]]
file3 = file3[["Customer", "State", "Gender", "Education", "Customer Lifetime Value", "Income", "Monthly Premium Auto", "Number of Open Complaints", "Policy Type", "Vehicle Class", "Total Claim Amount"]]              
```

```python
file1 = file1.rename(columns = {"ST": "State", "GENDER": "Gender"})
file2 = file2.rename(columns = {"ST": "State", "GENDER": "Gender"})
```

## 4. Merge the data frames

```python
column_names = file3.columns
column_names
```

```python
data = pd.DataFrame(columns = column_names)
data.head()
```

```python
data = pd.concat([data, file1, file2, file3], axis = 0)
```

```python
data.shape
```

## 5. Standardize the header names

```python
cols = []
for i in range(len(data.columns)):
    cols.append(data.columns[i].lower().replace(" ", "_"))
```

```python
data.columns = cols
data.head()
```

## 6. Examine the columns and look for empty cells

```python
data.dtypes
```

```python
data.info()
```

## 7. Reset the index

```python
data.reset_index(drop = True, inplace = True)
data.info()
```

## 8. Drop irrevelant columns

* Education
* Number of open complaints

```python
data = data.drop(columns = ["education", "number_of_open_complaints"], axis = 1)
```

## 9. Correct the values in the column 'Customer Lifetime Value'

```python
# Remove the % sign if applicable
data["customer_lifetime_value"] = data["customer_lifetime_value"].str.replace("%", "")
```

```python
# Change dtype to numerical
data["customer_lifetime_value"] = pd.to_numeric(data["customer_lifetime_value"], errors = "coerce")

# Multiply by 100
data["customer_lifetime_value"] = list(map(lambda x: (x * 100), data["customer_lifetime_value"]))
```

```python
data.head()
```

## 10. Remove duplicate rows

```python
data = data.drop_duplicates()
```

```python
data.shape
```

## 11. Filter out the data for customer with an income of 0 or less

```python
data = data[(data["income"] > 0)]

data.reset_index(drop = True, inplace = True)
data.info()
```

```python
data.head()
```
