### 3. My Holiday Preferences

```python
season = ""
destination = ""

def preferences(season, destination):
    
    # First question
    print("Which time of the year do you prefer to travel? ")
    
    while season.lower() not in {"spring", "summer", "fall", "winter"}:
            season = input("Please choose spring, summer, fall or winter. ")
            season.lower in {"spring", "summer", "fall", "winter"}
    print(f"I prefer to travel in {season}.")

    # Second question
    print("What is your favorite type of holiday destination? ")
    
    while destination.lower() not in {"city", "mountains", "ocean", "other"}:
            destination = input("Please choose city, mountains, ocean or type 'other'. ")
            destination.lower in {"city", "mountains", "ocean", "other"}
    
    if destination in {"city", "mountains", "ocean"}:
        print(f"I like the {destination} as a holiday destination the most.")
        
    else:
        other_answer = input("You don't like any of the suggested options? Please tell me what type of destination you prefer instead. ")
        print(other_answer)

    return

preferences(season, destination)
```
