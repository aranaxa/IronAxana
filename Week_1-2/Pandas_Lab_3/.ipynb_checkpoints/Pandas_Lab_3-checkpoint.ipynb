## 5. - Show a plot of the distribution of the `Total Claim Amount`, broken down by `Response` type. Try a boxplot and distribution plot, for each `Response` type. For the distribution plot, try to plot both kinds of `Response`s in one chart (seaborn's histplot, using the 'hue' parameter is very convenient here).

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
sns.boxplot(x = "Total Claim Amount", y = "Response", data = data)
```

```python
sns.histplot(data = data, x = "Total Claim Amount", hue = "Response")
```

```python
sns.boxplot(x = "Response", y = "Income", data = data)
```

```python
sns.histplot(data = data, x = "Income", hue = "Response")
```
